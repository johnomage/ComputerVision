```python
import cv2 as cv, numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv.imread("c:/opencv/images/input.jpg")

histogram = cv.calcHist([image], [0], None, [256], [0,256])

plt.hist(image.ravel(), 256, [0,256]); plt.show() #raven() flattens image array

# viewing separate color channels
color = ("b", "g", "r")

# separate colors and plot histogram
for i, col in enumerate(color):
    histogram2 = cv.calcHist([image], [i], None, [256], [0,256])
    plt.plot(histogram2, color=col)
    plt.xlim([0, 255])
    plt.xlabel("Pixels")
    
plt.show()
```
